# Statistics of dataset

```python
for col in num_col.columns:
    print(f"Skewness for {col}:{skew(df[col])}")
    print(f"Kurtosis for {col}:{kurtosis(df[col])}")
    print("-"*50)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
inertia = []
for k in range(1,11) :
    kmeans = KMeans(n_clusters = k , random_state = 42) 
    # creating a kmodel with k clusters and random state = 42 to ensure reproducibility
    kmeans.fit(num_col)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,11),inertia,marker="o")
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()   
    
```

```python
kmeans=KMeans(n_clusters=4,random_state=42)
df["Cluster"]=kmeans.fit_predict(num_col)
print(df.head())
```

```python
print("Cluster Centers :\n",kmeans.cluster_centers_)
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(df)
# create a scatter plot in 2D
plt.scatter(reduced[:, 0], reduced[:, 1], c=df['Cluster'], cmap='viridis')
plt.title("K-Means Clusters (k=4)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label='Cluster')
plt.show()
```

```python
pca = PCA(n_components=3)
reduced = pca.fit_transform(df)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(reduced[:, 0], reduced[:, 1], reduced[:, 2],
                c=df['Cluster'], cmap='viridis', s=50)

ax.set_title("3D Visualization of K-Means Clusters (k=4)")
ax.set_xlabel("PCA 1")
ax.set_ylabel("PCA 2")
ax.set_zlabel("PCA 3")
plt.colorbar(sc, label="Cluster")
plt.show()
```
